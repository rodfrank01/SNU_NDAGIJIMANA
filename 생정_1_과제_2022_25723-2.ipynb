**Step 6. Shannon Entropy**

```
import math
from collections import Counter

def entropy_function(dna_sequence):
    bases = Counter(dna_sequence)
    total_bases = sum(bases.values())
    
    entropy_value = 0.0
    for count in bases.values():
        probability = count / total_bases
        entropy_value -= probability * math.log2(probability)

    return entropy_value
```

```
pileup_1000['entropy'] = pileup_1000['matches'].map(lambda x : entropy_function(x))
pileup_1000
```
