**Step 5. Read Depth**

```
pileup_100.head()
```

```
# In the sequences, one can see unwanted symbols such as $,...
# Need to clean this

import re
toremove = re.compile(r'[<>$*#+-]|\^.')
pileup_100['matches'] = pileup_100['basereads'].apply(lambda x: toremove.sub('', x))

pileup_100.head(3)
```

```
# check the chrom, positions and matches information 
# we subset the data frame using position, chromosome and pure sequence (match) information

pileup_100[['chrom', 'pos', 'matches']] # now we have no unnecessary symbols 
pileup_100.to_csv('pileup_100.csv') # save the file 
```

```
pileup_100 = pileup_100[['chrom', 'pos', 'matches']] 
# we subset the data frame using position, chromosome and pure sequence (match) information
pileup_100['depth'] = pileup_100['matches'].str.len()
pileup_100.shape
# in the matches column, count the number of letters( the read depth)

pileup_100.head(3)

# Now we know the read depth at each position 
```

```
pileup_100.shape
```

**Step 6. Shannon Entropy**

```
import math
from collections import Counter

def entropy_function(dna_sequence):
    bases = Counter(dna_sequence)
    total_bases = sum(bases.values())
    
    entropy_value = 0.0
    for count in bases.values():
        probability = count / total_bases
        entropy_value -= probability * math.log2(probability)

    return entropy_value
```

```
pileup_100['entropy'] = pileup_100['matches'].map(lambda x : entropy_function(x))
pileup_100.head()
```

```
pileup_100['end'] = pileup_100['pos'] +  pileup_100['matches'].apply(len)
pileup_100['start+1'] = pileup_100['pos'] + 1
pileup_100
```

```
# filtering 2: positioning 
# the lengths of sequences are different, hard to draw a weblogo 
# filtering by position allows to set a similar range 
bed_pileup = pileup_100[['chrom','pos','end','entropy']]
bed_pileup.columns = ['chromosome','start','end', 'entropy' ]

bed_pileup['-12'] = bed_pileup['start'] - 12
bed_pileup['+11'] = bed_pileup['start'] + 11
bed_pileup.head()

pileup_fromstart = bed_pileup[['chromosome','-12','+11','entropy']]
```

```
y=bed_pileup[['chromosome','-12','+11','entropy']]
y.head()

y.to_csv('final.bed',header=None, index=False, sep='\t')
```
